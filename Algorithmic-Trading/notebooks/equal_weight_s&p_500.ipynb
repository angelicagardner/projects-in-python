```python
from datetime import datetime, timedelta
import math

import numpy as np
import pandas as pd
import requests
import yfinance as yf
```

```python
sp500 = pd.read_csv('../data/flat-ui__data-MonJun102024.csv')
```

```python
sp500
```

```python
sp500['Symbol'] = sp500['Symbol'].str.replace('.', '-')
symbols_list = sp500['Symbol'].unique().tolist()
symbols_list
```

```python
tickers = []
for ticker in symbols_list:
    try:
        ticker_info = yf.Ticker(ticker).info
        if 'marketCap' in ticker_info and ticker_info['marketCap'] is not None:
            tickers.append(ticker)
        else:
            print(f"No valid market cap data for {ticker}")
    except Exception as e:
        print(f"Ticker {ticker} is invalid or delisted: {e}")
```

```python
end_date = datetime.now() - timedelta(days=3)
end_date_pd = pd.to_datetime(end_date)
```

```python
start_date = end_date_pd - pd.DateOffset(years=2)
start_date
```

```python
df = yf.download(tickers=tickers, start=start_date, end=end_date_pd)
df = df.dropna()
df
```

```python
df = df.stack(future_stack=True)
df.index.names = ['date', 'ticker']
df.columns = df.columns.str.lower()
df
```
